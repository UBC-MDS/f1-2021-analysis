We will be visualizing Formula 1  data having multiple datasets for the year 2021. We will be looking at 5 datasets namely: 
- The `formula1_2021season_drivers.csv` which provides personal information and their all time best stats and records in formula 1.
- The `formula1_2021season_calendar.csv` contains information on each of the 22 tracks used in 2021. Information about lap records, track dimensions and information.
- The `formula1_2021season_raceResults.csv` includes information about the final race positions, lap times and points of each f1 driver in all the races that happened in 2021. 
- The `formula1_2021season_sprintQualifyingResults.csv` includes information about the final race positions, lap times and points of each f1 driver in all the races that happened in 2021. 
- The `formula1_2021season_teams.csv` includes information about the 10 teams that participated in f1 2021 season. 

Lets do some analysis on the race results 

```python
final_race_results_df = pd.read_csv("../data/formula1_2021season_raceResults.csv")
sprint_race_results_df = pd.read_csv("../data/formula1_2021season_sprintQualifyingResults.csv")
```

```python
final_race_driver_points = final_race_results_df.groupby(['Driver', 'Team'])['Points'].sum()
sprint_race_driver_points = sprint_race_results_df.groupby(['Driver', 'Team'])['Points'].sum()
driver_points = (final_race_driver_points + sprint_race_driver_points).sort_values(ascending=False).reset_index()
driver_points.index = [i for i in range(1,22)]
```

```python
driver_points
```

We can see that Max Verstappen is the champion of 2021 drivers cup, beating Lewis Hamilton by 8 points

```python
final_race_team_points = final_race_results_df.groupby(['Team'])['Points'].sum()
sprint_race_team_points = sprint_race_results_df.groupby(['Team'])['Points'].sum()
(final_race_team_points + sprint_race_team_points).sort_values(ascending=False).reset_index()
team_points = (final_race_team_points + sprint_race_team_points).sort_values(ascending=False).reset_index()
team_points.index = [i for i in range(1, 11)]
```

```python
team_points
```

In constrast, the constructor cup was won by Mercedes, who was ahead of RedBull by 28 points
